```python
#SETS

s={"KAni","Jani","Kani"}
print(s)
#True and 1 are same and false and 0 are same in sets
s1={2,True,False,5,1}  #Unordered, unindexed, no duplicate values
print(s1)
s1.add(10)
print(s1)

#Union of sets
s2={1,2,3,4,5}
s3={4,5,6,7,8}
s4=s3.union(s2)
print(s4)

#OR
s4=s2|s3
print(s4)

#In case of emoty sets we have to use set() function instead of {} because {} is used for creating empty dictionary
s5=set()

#Intersection of sets
#This can be done through intersection() or & operator

s6=set()
s7=set()
for i in range(6):
    s6.add(i)
for i in range(2,9):
    s7.add(i)
s8=s6.intersection(s7)
print(s8)

#OR
s8=s6&s7
print(s8)


#DOUBT
#Difference of sets
s9=s6.difference(s7)
print(s9)
#or
s9=s6-s7
print(s9)
```

```python
def function():
    for i in range(5):
        print("😊")

#Positional functions
def fun2(n1,n2):
    return (n1*n2)

def fun3(*name):
    print(name)
    print(name[2])
    

function()
print(fun2(5, 10))
print(fun2(2, 3))
fun3("Kanishka","Joshi","Darshil","Ameta")

#Not positioned functions
def fun4(City1,city2):
    print("I am from",City1)
    print("I am from",city2)

fun4(city2="Ahmedabad",City1="Surat")


def fun5 (**city):
    print(city)

fun5(city1="Surat",city2="Ahmedabad",city3="Vadodara")

#Default function
def fun6(name="Kanishka"):
    print("My name is",name)    

fun6()
fun6("Darshil")


def fun7(sub):
    for i in sub:
        print(i)
sub=["DSA","AI"]
fun7(sub)

#To specify the only the keyword arguments can be written so add"*"ArithmeticError"

#Keyword Arguments
def dob(*,n1):
    print(n1*2)
dob(n1=3)

#Positional Arguments
def dob(n1,/):
    print(n1*2)
dob(2)

#recurssion
def rec(K):
    if K>0:
        print(K)
        K-=1
        rec(K)
    else:
        print(K)

rec(-1)
    
def my_function(a,b,/,*,c,d):
    print("a=",a)
    print("b=",b)
    print("c=",c)
    print("d=",d)
    print(a+b+c+d)

my_function(5,6,c=2,d=4)
```
