4) Write a Python program to calculate the total, highest value, and lowest value from a tuple of integers without using the built-in sum(), max(), and min() functions

```python
t1 = (5, 8, 2, 10, 1)

total = 0
largest = t1[0]
smallest = t1[0]

for num in t1:
    total += num

    if num > largest:
        largest = num

    if num < smallest:
        smallest = num

print("Total =", total)
print("Largest =", largest)
print("Smallest =", smallest)
```
