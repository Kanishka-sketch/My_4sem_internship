```python
#Q1. Write a Python function to find the maximum of three numbers.

def maximum(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c


print(maximum(10, 25, 15))  
```

```python
#2. Write a Python function that takes a list and returns a new list with distinct elements from the first list.

def distinct_list(lst):
    new_list = []

    for item in lst:
        if item not in new_list:
            new_list.append(item)

    return new_list

# Example usage
numbers = [1, 2, 2, 3, 4, 4, 5]
print(distinct_list(numbers))
```

```python
#3. Write a Python function to multiply all the numbers in a list.

def multiply(a):
    result = 1

    for i in a:
        result = result * i

    return result


x = [2, 3, 4]

print(multiply(x))
```

```python
#4. Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument.

def factorial(n):
    fact= 1

    for i in range(1, n + 1):
        fact = fact * i

    return fact



print(factorial(5))
```

```python
#5. Write a Python program to reverse a string.

def reverse_string(text):
    return text[::-1]


print(reverse_string("Python"))
```

```python
#6. Write a Python function to check whether a number falls within a given range.

def check(num):
    if num >= 10 and num <= 20:
        print(f"{num} is in range")
    else:
        print(f"{num} is not in range")

check(15)
```

```python
#7. Write a Python function to Print Even Numbers from a Given List
def even_numbers(a):
    for i in a:
        if i % 2 == 0:
            print(i)

x = [1, 2, 3, 4, 5, 6]

even_numbers(x)
```

```python
#8. Write a Python function that takes a number as a parameter and checks whether the number is prime or not.
def check_prime(num):
    count = 0

    for i in range(1, num + 1):
        if num % i == 0:
            count = count + 1

    if count == 2:
        print("Prime Number")
    else:
        print("Not Prime Number")



check_prime(12)
```

```python
#9. Write a Python function that accepts a string and counts the number of upper and lower case letters.

def count_letters(text):
    upper = 0
    lower = 0

    for i in text:
        if i.isupper():
            upper = upper + 1

        elif i.islower():
            lower = lower + 1

    print("Upper case letters:", upper)
    print("Lower case letters:", lower)

count_letters("PyThOn")
```
