Regular Expression

```python
import re

text = "Python"

pattern = r'^[A-Za-z]+$'

if re.match(pattern, text):
    print("Only alphabets")
else:
    print("Contains other characters")
```

```python
import re

mobile = "9876543210"

pattern = r'^[6-9][0-9]{9}$'

if re.match(pattern, mobile):
    print("Valid Mobile Number")
else:
    print("Invalid Mobile Number")
```

2. DATETIME FUNCTIONS IN PANDAS

```python
import pandas as pd

data = {
    'Date': ['2025-01-10', '2025-02-15', '2025-03-20']
}

df = pd.DataFrame(data)

print(df)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
print(type(df))
```

Extract Year,month, month_name, day ,day_name,quarter

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()
df['Day'] = df['Date'].dt.day
df['Day_Name'] = df['Date'].dt.day_name()
df['Quarter'] = df['Date'].dt.quarter

print(df)
```

3. CSV FILE DATA CLEANING & ANALYSIS

What is Data Cleaning?

Data cleaning means:

removing missing values
fixing wrong data
removing duplicates
formatting data properly

```python
df = pd.read_csv('Anime.csv')
print(df)
```

```python
#Check missing values
print(df.isnull().sum())
```

```python
#Fill Missing Values
df['Rating'].fillna(df['Rating'].mean(), inplace=True)
```

```python
#Find Top Rated Anime
top_anime = df.sort_values('Rating', ascending=False)

print(top_anime[['Name', 'Rating']].head(10))
```

```python
#Create Pivot
pivot = pd.pivot_table(
    df,
    values='Rating',
    index='Type',
    aggfunc='mean'
)

print(pivot)
```
