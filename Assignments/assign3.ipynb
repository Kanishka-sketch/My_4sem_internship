```python
#Q1.1) Practise Dictionary Tuple Set

#Dictionary

dict1={"Name":"Kanishka","Age":20,"City":"Chittorgarh"}
print(dict1)
#Accessing Values
print(dict1["Name"])
print(dict1["Age"])
#Updating values
dict1["Age"]=21
print(dict1)
#Adding new key-value pair
dict1["Role"]="Student"
print(dict1)
#Removing key-value pair
dict1.pop("Role")
print(dict1)

#TUPLE
tuple1=("Kanishka",20,"Chittorgarh","Kanishka")
print(tuple1)
#Accessing values
print(tuple1[0])
print(tuple1[1])
#Tuples are immutable, so we cannot update or add values
#Tuple methods
print(tuple1.count("Kanishka"))
print(tuple1.index(20))
#tuple1.delete(tuple1)  # Tuples are immutable

#SET=stores unique values and is unordered
set1={"Kanishka",20,"Chittorgarh"}
print(set1)
#Duplicate Values Removed Automatically
set2={"Kanishka",20,"Chittorgarh","Kanishka"}
print(set2)
#Adding new element to set
set2.add("Student")
print(set2)
#Removing element from set
set2.remove("Student")
print(set2)
```

```python
#Q3.Write a program to check Palindrome Number
num = input("Write the no to check : ")

b = ""

for i in range(len(num)-1, -1, -1):
    b = b + num[i]

if b == num:
    print("Palindrome")
else:
    print("Not Palindrome")
```
