3) Replace NaN values with average of columns 

```python
# Create array
arr = np.array([[10, 20, np.nan],
                [40, np.nan, 60],
                [70, 80, 90]])
print("Original:")
print(arr)
```

np.mean() propagates NaN values, while np.nanmean() ignores them during average calculation.

```python

# Column mean ignoring NaN
#Axis = 0 , column wise mean 
mean = np.nanmean(arr, axis=0)

# Replace NaN
arr[np.isnan(arr)] = mean[np.where(np.isnan(arr))[1]]
# 1= give SECOND array , 0 = give FIRST array
print(arr)
```

4) Replace negative value with zero in numpy array using replace

there is NO replace() function used.

This method is called:

Boolean Indexing / Conditional Replacement

Then Why Did Your Question Say “using replace”?

Because NumPy does not have a direct .replace() method like Pandas.

In NumPy, replacement is usually done using:

arr[condition] = value

BEST PRACTICE is this

```python
arr = np.array([10,-5,20,-8,30])

arr[arr < 0] = 0

print(arr)
```

np.where(condition, true_value, false_value)

if condition is True → use true_value
else → use false_value

```python
#OR(ALTERNATIVE)
arr = np.array([10,-5,20,-8,30])

arr = np.where(arr < 0, 0, arr)

print(arr)
```

In Pandas

There is an actual replace() function.

```python
import pandas as pd

s = pd.Series([10,-5,20,-8])

s.replace(-5,0)
```
