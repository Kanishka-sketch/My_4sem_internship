```python
#1) Replace Nan with 0 and Interchange 3 rows and 3 columns of 2D array [[6, -8, 73, -110], [np.nan, -8, 0, 94]]

import numpy as np
arr = np.array( [[6, -8, 73, -110], 
                 [np.nan, -8, 0, 94]])
print("2D_Array:\n",arr)

replace = np.nan_to_num(arr,nan = 0)
print("Replace Nan with 0:")
print(replace)

print("Interchanged:")
arr[:,[0,3]] = arr[:,[3,0]]
print(arr)
```

```python
#2) Move axes of 3D array to new positions
import numpy as np

arr = np.arange(24).reshape(2,3,4)

print("Original Shape:\n", arr,"\n",arr.shape)

new_arr = np.moveaxis(arr, 0, 2)

print("New Shape:",new_arr, new_arr.shape)
```

```python
#3) Replace NaN values with average of columns 
import pandas as pd
data = {
    "Math": [90, 80, np.nan, 70],
    "Science": [85, np.nan, 75, 95]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

df.fillna(df.mean(), inplace=True)
print("\nDataFrame after Replacing NaN with Column Average:")
print(df)
```

```python
#4) Replace negative value with zero in numpy array using replace 
arr = np.array([1,-1,2,-2,3,4])
print("Original Array:\n",arr)

arr[arr < 0] = 0
print("Replaced array:\n",arr)
```

```python
#6) Solve the following equation using linalg() and inverse Matrix method x - 2y + 3z = 9 -x + 3y - z = -6 2x - 5y + 5z = 17 7) Using "Customizing Matplotlib Visualizations" discussed in Numpy session4 notes plot graph to compare your at least 2 semester result....

#Method 1: Using numpy.linalg.solve()
import numpy as np

A = np.array([
    [1, -2, 3],
    [-1, 3, -1],
    [2, -5, 5]
])

B = np.array([9, -6, 17])

X = np.linalg.solve(A, B)

print("Solution:")
print("x =", X[0])
print("y =", X[1])
print("z =", X[2])

#Method 2: Using Inverse Matrix
A = np.array([
    [1, -2, 3],
    [-1, 3, -1],
    [2, -5, 5]
])

B = np.array([9, -6, 17])

A_inv = np.linalg.inv(A)

X = np.dot(A_inv, B)

print("Solution:")
print("x =", X[0])
print("y =", X[1])
print("z =", X[2])
```

```python
import matplotlib.pyplot as plt

subjects = ["Maths", "Python", "DBMS", "OS", "CN"]

sem1 = [75, 80, 78, 72, 85]
sem2 = [82, 88, 84, 79, 90]

plt.figure(figsize=(8,5))

plt.plot(subjects, sem1,
         marker='o',
         linewidth=2,
         label='Semester 1')

plt.plot(subjects, sem2,
         marker='s',
         linewidth=2,
         label='Semester 2')

plt.title("Semester Result Comparison")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(True)
plt.legend()

plt.show()
```

```python
#Bar Chart

subjects = ["Maths", "Python", "DBMS", "OS", "CN"]

sem1 = [75, 80, 78, 72, 85]
sem2 = [82, 88, 84, 79, 90]

x = np.arange(len(subjects))
width = 0.35

plt.figure(figsize=(8,5))

plt.bar(x-width/2, sem1, width,
        label="Semester 1")

plt.bar(x+width/2, sem2, width,
        label="Semester 2")

plt.xticks(x, subjects)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Semester Results Comparison")
plt.legend()
plt.grid(axis='y')

plt.show()
```
