```python
#1.Study the open weather API show more data in your API calling program

# API CALLING 
import requests

def weather_data(city):
    url=f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid=345b7307ead3b76a95697f5fc9adab3d&units=metric"
    
    try:
        response=requests.get(url)
        response.raise_for_status()  
        data=response.json()
        print(f"Weather data for {city}:")
        count=0
        for key, value in data['main'].items():
            print(f"{key}: {value}")
            count += 1
            if count > 6:
                break

    except requests.exceptions.RequestException as e:
        print(f"Error fetching weather data for {city}: {e}")

city=input("Enter the city name: ")
weather_data(city)
```

```python
#2.Rock Paper Scissors
import random

# Initialize scores
uscore = 0  # User score
cscore = 0  # Computer score

# List of possible choices
choices = ["rock", "paper", "scissor"]

print("Welcome to Rock-Paper-Scissors Game!")
print("Rules:")
print("Rock beats Scissor, Scissor beats Paper, Paper beats Rock")
print("You get 2 points for winning, 1 point each for a tie, 0 for losing\n")

# Loop to play multiple rounds
while True:
    # User input
    uch = input("Enter your choice (rock/paper/scissor): ").lower()
    
    # Check valid input
    if uch not in choices:
        print("Invalid choice! Try again.\n")
        continue
    
    # Computer choice
    cch = random.choice(choices)
    print(f"Computer chose: {cch}")
    
    # Game logic
    if uch == cch:
        # Tie
        uscore += 1
        cscore += 1
        print("It's a Tie! Both get 1 point.")
    
    elif ((uch == "rock" and cch == "scissor") or
          (uch == "paper" and cch == "rock") or
          (uch == "scissor" and cch == "paper")):
        # User wins
        uscore += 2
        print("You Win! +2 points")
    
    else:
        # Computer wins
        cscore += 2
        print("Computer Wins! +2 points to Computer")
    
    # Display current scores
    print(f"Current Score -> You: {uscore} | Computer: {cscore}\n")
    
    # Ask if user wants to play again
    again = input("Do you want to play another round? (yes/no): ").lower()
    if again == "no":
        break

# Final result
print("\nFinal Score:")
print(f"You: {uscore} | Computer: {cscore}")

if uscore > cscore:
    print("Congratulations! You are the overall winner!")
elif uscore < cscore:
    print("Computer wins the game! Better luck next time.")
else:
    print("The game ends in a tie!")
```
