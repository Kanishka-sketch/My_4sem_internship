```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
spark = SparkSession.builder \
    .appName("Sales Analysis") \
    .getOrCreate()
```

```python
df = spark.read.csv(
    "data/sales.csv",
    header=True,
    inferSchema=True
)

print("\nOriginal Data")
df.show()
```

```python
print("Products Sorted by Sales (Descending)")
sorted_df = df.orderBy(col("sales").desc())
sorted_df.show()
```

```python
print("Top 3 Products by Sales")
top3_df = sorted_df.limit(3)
top3_df.show()
```

```python
print("Products with Sales Greater Than 80000")
high_sales_df = df.filter(col("sales") > 80000)
high_sales_df.show()
```

```python
high_sales_df.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("output/high_sales_products")

print("Output saved successfully.")
```

```python
spark.stop()
```
