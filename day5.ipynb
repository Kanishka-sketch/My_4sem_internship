```python
#Python CSV

import csv

d=[
    ["Name","Branch","Age"],
    ["Kani","D.S",20],
    ["Dishu","IT",21]
]

with open("Data.csv","w",newline='') as file:        #newline=''......it is used to remove the added extra line in the csv file created ,after each line.
    writer = csv.writer(file)
    for row in d:
        writer.writerow(row)

#Append the data

d=[
    ["Div","D.S",20],
    ["Dibbi","D.S",20]
]

with open("Data.csv",'a',newline='')as file:
    writer= csv.writer(file)
    for row in d:
        writer.writerow(row)



#Read the csv file .....
with open("Data.csv","r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)

#Accessing Particular Columns
with open("Data.csv","r") as file:
    reader= csv.reader(file)

    for row in reader:
        print(row[0])
```
