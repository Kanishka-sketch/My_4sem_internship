Transpose and Concatenation

```python
import numpy as np 
#For single dim, axis= 0 , no 1

a = np.array([1,2,3,4])
b = np.array([7,8,9,10])
 
res1 = np.transpose((a,b))

print(res1)
res = np.concatenate((a,b) , axis = 0)

print(res)

```

```python
c = np.array([[1,2] , [3,4]])
d = np.array([[6,7] , [8,2]])

print(c)
print(d)

print("Moduling the two variables:")
print(c%d)
print("Power operation.")
print(c**d)
```
