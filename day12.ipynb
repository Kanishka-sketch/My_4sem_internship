Transpose and Concatenation

```python
import numpy as np 
#For single dim, axis= 0 , no 1

a = np.array([1,2,3,4])
b = np.array([7,8,9,10])
 
res1 = np.transpose((a,b))

print(res1)
res = np.concatenate((a,b) , axis = 0)

print(res)

```

```python
c = np.array([[1,2] , [3,4]])
d = np.array([[6,7] , [8,2]])

print(c)
print(d)

print("Moduling the two variables:")
print(c%d)
print("Power operation.")
print(c**d)
```

```python
#Missing values

d = np.array([1,2,np.nan,3,np.nan])
res = np.isnan(d)
print(d)
print(res)
```

```python
res1 = (~np.isnan(d))
print(res1)

#Replacing nan values
res2 = np.nan_to_num(d,nan = 0)
print(res2)
```

```python
# 3x + 2y = 5
# x + 2y = 5
import numpy as np
arr1 = np.array([       #left side value
    [3,2],[1,2]
])
arr2 = np.array([5,5])  #Right side value

res = np.linalg.solve(arr1,arr2)
print(res)   #IT gives the  x = 0 ,y = 2.5

#4x +2y = 8
# 5x +3y =9
arr1 = np.array([       #left side value
    [4,2],[5,3]
])
arr2 = np.array([8,9])  #Right side value

res = np.linalg.solve(arr1,arr2)
print(res)  
```

```python
arr1 = np.array([       #left side value
    [3,2],[1,2]
])
arr2 = np.array([5,5])  #Right side value

arr_inv = np.linalg.inv(arr1)     #inverse of the array eq .
res = np.dot(arr_inv , arr2)
print(res)  
```

```python
#100 evenly spaved point bw-0 and 10
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)          #we will get the sine values of the array
plt.plot(x,y,label = 'sine(x)',color = 'red',linestyle = "--")
plt.title("Sine Wave")
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

Scatter plot

```python
x = np.random.rand(50)
y= np.random.rand(50)
plt.scatter(x,y,label = 'sine(x)',color = 'red',alpha = 0.7)
plt.title("Random Values")
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

Bar plot

```python
catagories = ["a","b","c","d"]
values = [1,2,3,4]

plt.bar(catagories,values,color =['red','blue','green','orange'])
plt.title("Bar plot")
plt.xlabel("catagories")
plt.ylabel("Values")
plt.show()
```

Histogram

```python
data = np.random.randn(1000)
plt.hist(data,bins=20, alpha = 0.7,color = 'blue')
plt.title("Histogram")
plt.xlabel("bins")
plt.ylabel("Frequency")
plt.show()
```

```python
#Pie chart
labels = ['Python','Cpp','Java']
sizes = [40,50,60]

plt.pie(sizes,labels = labels,autopct = '%1.1f%%' , startangle = 90)
plt.title("Pie chart")
plt.show()
```
