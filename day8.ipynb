```python
import pandas as pd
df = pd.DataFrame({'A':[1,2,3,4,5,] , 'B':[10,20,30,40,50]})
print(df)

print("Addition:\n",df+10)
print("\nSubstraction:\n",df-2)
print("\nMultiplication:\n",df*10)
```

```python
#At a particular row 

print("\nAddition:\n",df['B']+4)
```

```python
#Operrations between two dataframes 

df1 = pd.DataFrame({'A':[1,2,3,4] , 'B':[10,20,30,40]})
df2 = pd.DataFrame({'A':[11,12,13],'B':[100,200,300]} ,index=[1,2,3])
print(df1)
print(df2)

print("Addition:\n",df1+df2)
print("\nSubtraction:\n",df1+df2)
print("Multiplication:\n",df1+df2)
```

```python

one = pd.DataFrame({
    'Name':["A1","B1","C1","D1"],
    "Sub":["Maths","Phy","Chem","Cpp"],
    "Marks":[30,50,47,55]
})

two = pd.DataFrame({
    'Name':["P1","B1","C1","Q1"],
    "Sub":["Python","Java","Chem","Cpp"],
    "Marks":[30,17,47,20]
},index=[1,2,3,4])

print("First dataframe:")
print(one)
print("\nSecond dataframe:")
print(two)

print("\nMerged dataframe")
res=pd.concat([one,two])    #Dataframe is always passed in the square brackets
print(res)

#Concat through keys..........
print("Concat through Keys.")
res1=pd.concat([one,two] , keys=["X","Y"])
print(res1)

#Ignore index concate......the given index will be ingnored and normal count will be initialised.for more clarification do compare the previous output.
res2=pd.concat([one,two],ignore_index=True)
print(res2)

#Use of Axis
#Axis=0....(Vertical output),Axis=1.....(Horizontal output)
res3=pd.concat([one,two],axis=1)
res4=pd.concat([one,two],axis=0)
print(res3)
print(res4)
```

```python
#Merginggggg....
one= pd.DataFrame({
    'id':[1,2,3,4],
    'Name':["A1","B1","C1","D1"],
    'Sub':["Maths","Phy","Chem","Cpp"],
})

two = pd.DataFrame({
    'id':[1,2,3,4],
    'Name':["P1","B1","C1","Q1"],
    'Sub':["Python","Java","Chem","Cpp"],
    'Marks':[30,17,47,20]
})
print("First dataframe:")
print(one)
print("\nSecond dataframe:")
print(two)

print("Merged dataframe:")
res= one.merge(two,on="id")
print(res)
```
