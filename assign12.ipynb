```python
#1) Create numpy array and perform following operation > Convert 1D array to 2D
a = np.array([1,2,3,4,5,6])
print("Original",a)
a2d = a.reshape(2,3)
print("Reshaped_2D:\n",a2d)
```

```python
#Print Array Attributes(Like shape, dimenssion, data type, itemsize)

print(a2d.shape) #shape
print(a2d.ndim)
print(a2d.dtype)
print(a2d.itemsize)
```

```python
#Replace all odd numbers in a NumPy array with -1 by itterating using for loop [[23, 56, 78, 93], [71, 82,13, 24]] 

import numpy as np
a8 = np.array([[23, 56, 78, 93], 
                [71, 82,13, 24]])

# for i in np.nditer(a8):
#     print(i);

for i in range(a8.shape[0]):
    for j in range(a8.shape[1]):
        if a8[i][j] %2 != 0:
            a8[i][j] = -1

print("Modofied:\n",a8)
```

```python
#Get the indices of non-zero elements in an array [1, 0, 2, 0, 3, 0, 4] 

a9 = np.array([1, 0, 2, 0, 3, 0, 4])
indices = np.nonzero(a9)
print(indices)
```
